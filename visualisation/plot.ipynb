```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

CSV_FOLDER = '../csv/tabula-muris_'
```

```python
# Read all files in the CSV folder as seperate dataframes
# Here we do all the 5-ways K-shots analysis
relationnet_fc_diff_mean_05_k = pd.read_csv(CSV_FOLDER + 'relationnet_fc_diff_mean_05_k.csv')
relationnet_l1_mean_05_k = pd.read_csv(CSV_FOLDER + 'relationnet_l1_mean_05_k.csv')
protonet_05_k = pd.read_csv(CSV_FOLDER + 'protonet_05_k.csv')
matchingnet_05_k = pd.read_csv(CSV_FOLDER + 'matchingnet_05_k.csv')
maml_05_k = pd.read_csv(CSV_FOLDER + 'maml_05_k.csv')
```

```python
relationnet_fc_diff_mean_05_k.head()
```

```python

# We make where the x axis is the number of shots and the y axis is the mean accuracy, with the standard deviation as error bars

ALPHA=0.1

# For the val split:
# RelationNet
relationnet_fc_diff_mean_05_k_val = relationnet_fc_diff_mean_05_k[relationnet_fc_diff_mean_05_k['split'] == 'val']
relationnet_fc_diff_mean_05_k_val = relationnet_fc_diff_mean_05_k_val.sort_values(by=['n_shot'])
relationnet_fc_diff_mean_05_k_val.head()
# plot it
plt.figure(figsize=(10, 5))
sns.set(style="darkgrid")
# We add the standard deviation as error bars filled between
sns.lineplot(x='n_shot', y='acc_mean', data=relationnet_fc_diff_mean_05_k_val, label='RelationNet')
plt.fill_between(relationnet_fc_diff_mean_05_k_val['n_shot'], relationnet_fc_diff_mean_05_k_val['acc_mean'] - relationnet_fc_diff_mean_05_k_val['acc_std'], relationnet_fc_diff_mean_05_k_val['acc_mean'] + relationnet_fc_diff_mean_05_k_val['acc_std'], alpha=ALPHA)

# RelationNet L1
relationnet_l1_mean_05_k_val = relationnet_l1_mean_05_k[relationnet_l1_mean_05_k['split'] == 'val']
relationnet_l1_mean_05_k_val = relationnet_l1_mean_05_k_val.sort_values(by=['n_shot'])
relationnet_l1_mean_05_k_val.head()
# plot it
sns.lineplot(x='n_shot', y='acc_mean', data=relationnet_l1_mean_05_k_val, label='RelationNet L1')
plt.fill_between(relationnet_l1_mean_05_k_val['n_shot'], relationnet_l1_mean_05_k_val['acc_mean'] - relationnet_l1_mean_05_k_val['acc_std'], relationnet_l1_mean_05_k_val['acc_mean'] + relationnet_l1_mean_05_k_val['acc_std'], alpha=ALPHA)

# ProtoNet
protonet_05_k_val = protonet_05_k[protonet_05_k['split'] == 'val']
protonet_05_k_val = protonet_05_k_val.sort_values(by=['n_shot'])
protonet_05_k_val.head()
# plot it
sns.lineplot(x='n_shot', y='acc_mean', data=protonet_05_k_val, label='ProtoNet')
plt.fill_between(protonet_05_k_val['n_shot'], protonet_05_k_val['acc_mean'] - protonet_05_k_val['acc_std'], protonet_05_k_val['acc_mean'] + protonet_05_k_val['acc_std'], alpha=ALPHA)

# MatchingNet
matchingnet_05_k_val = matchingnet_05_k[matchingnet_05_k['split'] == 'val']
matchingnet_05_k_val = matchingnet_05_k_val.sort_values(by=['n_shot'])
matchingnet_05_k_val.head()
# plot it
sns.lineplot(x='n_shot', y='acc_mean', data=matchingnet_05_k_val, label='MatchingNet')
plt.fill_between(matchingnet_05_k_val['n_shot'], matchingnet_05_k_val['acc_mean'] - matchingnet_05_k_val['acc_std'], matchingnet_05_k_val['acc_mean'] + matchingnet_05_k_val['acc_std'], alpha=ALPHA)

# MAML
maml_05_k_val = maml_05_k[maml_05_k['split'] == 'val']
maml_05_k_val = maml_05_k_val.sort_values(by=['n_shot'])
maml_05_k_val.head()
# plot it
sns.lineplot(x='n_shot', y='acc_mean', data=maml_05_k_val, label='MAML')
plt.fill_between(maml_05_k_val['n_shot'], maml_05_k_val['acc_mean'] - maml_05_k_val['acc_std'], maml_05_k_val['acc_mean'] + maml_05_k_val['acc_std'], alpha=ALPHA)

plt.ylabel('Mean accuracy')
plt.xlabel('Number of shots')
plt.title('5-way K-shot validation accuracy')
plt.savefig('../plots/5-way_K-shot_validation_accuracy.png', dpi=300)
plt.show()
```
